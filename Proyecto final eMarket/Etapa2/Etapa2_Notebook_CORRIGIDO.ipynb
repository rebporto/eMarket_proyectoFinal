```python

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.boxplot(y=productos_detalle['PrecioUnitario'])
plt.title("Boxplot de PrecioUnitario (con outliers)")
plt.show()

q1 = productos_detalle['PrecioUnitario'].quantile(0.25)
q3 = productos_detalle['PrecioUnitario'].quantile(0.75)
iqr = q3 - q1
filtro = (productos_detalle['PrecioUnitario'] >= (q1 - 1.5 * iqr)) & (productos_detalle['PrecioUnitario'] <= (q3 + 1.5 * iqr))
productos_detalle_limpio = productos_detalle[filtro]

sns.boxplot(y=productos_detalle_limpio['PrecioUnitario'])
plt.title("Boxplot de PrecioUnitario (sin outliers)")
plt.show()
```

### Porcentaje de Nulos por Columna – Antes y Después  
Visualización del impacto de la limpieza aplicada al dataset `clientes.csv`. Se eliminaron columnas con alto porcentaje de nulos (como "Fax"), mientras que otras como "Código Postal" se mantuvieron para evaluación futura.


```python
clientes = pd.read_csv("clientes.csv")  # ajustar caminho se necessário
clientes_limpio = clientes.drop(columns=["Fax", "Region"], errors='ignore')

nulos_antes = clientes.isnull().mean() * 100
nulos_despues = clientes_limpio.isnull().mean() * 100

comparativo = pd.DataFrame({"Antes": nulos_antes, "Después": nulos_despues}).fillna(0)
comparativo = comparativo[comparativo["Antes"] > 0]

fig, ax = plt.subplots(figsize=(8, 4))
x = range(len(comparativo))
ax.bar(x, comparativo["Antes"], width=0.4, label="Antes")
ax.bar([i + 0.4 for i in x], comparativo["Después"], width=0.4, label="Después")
ax.set_xticks([i + 0.2 for i in x])
ax.set_xticklabels(comparativo.index, rotation=45)
ax.set_ylabel("% de valores nulos")
ax.set_title("Porcentaje de Nulos por Columna - Antes y Después")
ax.legend()
plt.tight_layout()
plt.show()
```

### Transformación de formato de fecha y creación de nueva variable  
Se convirtió la columna `FechaFactura` a formato datetime, y se creó la variable `DiasEntrega` calculando la diferencia entre la fecha de entrega y la fecha de emisión.


```python
facturas = pd.read_csv("facturas.csv") 

facturas["FechaFactura"] = pd.to_datetime(facturas["FechaFactura"], errors="coerce")
facturas["FechaEnvio"] = pd.to_datetime(facturas["FechaEnvio"], errors="coerce")
facturas["DiasEntrega"] = (facturas["FechaEnvio"] - facturas["FechaFactura"]).dt.days

facturas[["FechaFactura", "FechaEnvio", "DiasEntrega"]].head()
```
